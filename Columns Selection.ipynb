```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])  #Logic kya lagaya ki initials jo hai naa sabke ko rakh liya,agar true hua toh woh lag jaayega nahi toh cities
```

```python
df['city'] = cities
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 360].index.tolist()
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df
```

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])                     #Kitna over ka kitna ball
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df                                                                
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')     #Kitni deliveries phenk chuke hai
df                                                                              
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)   #Kitne out huye hai
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']       #Run Rate Nikalne ka Formula
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
df['last_five'] = last_five
```

```python
df                 #Jaahir si baat hai pehle 5 overs mai aise hi hoga
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
final_df
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])         #Shuffle kar diya
```

```python
final_df.sample(2)
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer            #yeh strings ko convert karne mai madad karega
```

```python
from sklearn.preprocessing import OneHotEncoder          #Transforming categorical data into some numbers or integers
```

```python
from sklearn.pipeline import Pipeline                    #A machine learning pipeline is a way to codify and automate the workflow it takes to produce a machine learning model. Machine learning pipelines consist of multiple sequential steps that do everything from data extraction and preprocessing to model training and deployment.
```

```python
from sklearn.preprocessing import StandardScaler                #yeh saare normal dependencies liye 
from sklearn.ensemble import RandomForestRegressor                  
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error,accuracy_score
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))        # It is pronounced as R squared and is also known as the coefficient of determination. It works by measuring the amount of variance in the predictions explained by the dataset.
```

```python
import joblib
joblib.dump(pipe, 'main.joblib')

```
